# load modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# load K4, K9 and K27

```python
bed_columns = ["chr", "start", "end"]
df_k4 = pd.read_table( "../../qbb2020/data/K4me3.bed", names=bed_columns )
df_k4
```

```python
bed_columns = ["chr", "start", "end"]
df_k9 = pd.read_table( "../../qbb2020/data/K9me3.bed", names=bed_columns )
df_k9
```

```python
bed_columns = ["chr", "start", "end"]
df_k27 = pd.read_table( "../../qbb2020/data/K27me3.bed", names=bed_columns )
df_k27
```

# Adding Widths to DataFrames

```python
df_k4.loc[:,"width"] = df_k4.loc[:,"end"] - df_k4.loc[:,"start"]
df_k4
```

```python
df_k9.loc[:,"width"] = df_k9.loc[:,"end"] - df_k9.loc[:,"start"]
df_k9
```

```python
df_k27.loc[:,"width"] = df_k27.loc[:,"end"] - df_k27.loc[:,"start"]
df_k27
```

# plot histogram of width

```python
fig, ax = plt.subplots( nrows=3 )
fig.set_figheight(15)
ax[0].hist( df_k4.loc[:,"width"] )
ax[1].hist( df_k9.loc[:,"width"] )
ax[2].hist( df_k27.loc[:,"width"] )
ax[0].tick_params(rotation=45)
ax[2].tick_params(rotation=45)
ax[0].set_xlabel("Width", fontsize=16)
ax[1].set_xlabel("Width", fontsize=16)
ax[2].set_xlabel("Width", fontsize=16)
ax[0].set_ylabel("Count", fontsize=16)
ax[1].set_ylabel("Count", fontsize=16)
ax[2].set_ylabel("Count", fontsize=16)
ax[0].set_title("Width distribution in K4", fontsize=20)
ax[1].set_title("Width distribution in K9", fontsize=20)
ax[2].set_title("Width distribution in K27", fontsize=20)
fig.tight_layout(pad=5.0)
```

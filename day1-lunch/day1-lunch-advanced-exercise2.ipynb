# load modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# load K4, K9 and K27

```python
bed_columns = ["chr", "start", "end"]
df_k4 = pd.read_table( "../../qbb2020/data/K4me3.bed", names=bed_columns )
df_k4
```

```python
bed_columns = ["chr", "start", "end"]
df_k9 = pd.read_table( "../../qbb2020/data/K9me3.bed", names=bed_columns )
df_k9
```

```python
bed_columns = ["chr", "start", "end"]
df_k27 = pd.read_table( "../../qbb2020/data/K27me3.bed", names=bed_columns )
df_k27
```

# Adding Widths to DataFrames

```python
df_k4.loc[:,"width"] = df_k4.loc[:,"end"] - df_k4.loc[:,"start"]
df_k4
```

```python
df_k9.loc[:,"width"] = df_k9.loc[:,"end"] - df_k9.loc[:,"start"]
df_k9
```

```python
df_k27.loc[:,"width"] = df_k27.loc[:,"end"] - df_k27.loc[:,"start"]
df_k27
```

# plot histogram of width

```python

plt.hist( np.log2(df_k4.loc[:,"width"]), bins = 10, density = True, alpha = 0.5, label = 'K4' )
plt.hist( np.log2(df_k9.loc[:,"width"]), bins = 10, density = True, alpha = 0.5, label = 'K9' )
plt.hist( np.log2(df_k27.loc[:,"width"]), bins = 10, density = True, alpha = 0.5, label = 'K27' )
plt.legend()
plt.xlabel("log2 of Width")
plt.ylabel("Normaized Probability")
plt.title("Normalized log2 Width Distribution of ChipSeq Data")
```
